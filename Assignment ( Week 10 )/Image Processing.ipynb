```python
# importing libraries

import numpy as np 
import pandas as pd
from matplotlib import pyplot as plt 
import cv2 as cv
from PIL import Image
```

### Exploratory data analysis (EDA)

```python
# to visualise the first image from the dataset

image1 = cv.imread("Dataset/1.jpg")
image1
```

```python
# to display the image in RGB Format

image_rgb1 = cv.cvtColor(image1, cv.COLOR_BGR2RGB)
plt.imshow(image_rgb1)
```

```python
# to visualise the last image from the dataset

image20 = cv.imread("Dataset/20.jpg")
image20
```

```python
# to display the image in RGB Format

image_rgb20 = cv.cvtColor(image20, cv.COLOR_BGR2RGB)
plt.imshow(image_rgb20)
```

```python
# To visualze the image in gray scale

image1gray=cv.cvtColor(image1,cv.COLOR_BGR2GRAY)
Image.fromarray(image1gray)
```

```python
# To visualze the image in gray scale

image20gray=cv.cvtColor(image20,cv.COLOR_BGR2GRAY)
Image.fromarray(image20gray)
```
