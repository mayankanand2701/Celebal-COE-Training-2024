## Visualization

### Histogram

```python
plt.figure(figsize=[12,6])
plt.hist(df["price"], bins=80, color="purple")
plt.xlabel("Price in USD")
plt.ylabel("Frequency")
plt.title("Diamonds Price Histogram")
plt.show()
```

### Line Chart

```python
# To visulize how the price changes with the carat

plt.figure(figsize=(12, 6))
plt.plot(df1['price'],df1['carat'])
plt.title('Carat v/s Price')
plt.xlabel('Price')
plt.ylabel('Carat')
plt.show()
```

```python
# To visulize how the price changes with the incrase in depth
plt.figure(figsize=(12, 6))
plt.plot(df1['price'],df1['depth'])
plt.title('Depth v/s Price')
plt.xlabel('Price')
plt.ylabel('Depth')
plt.show()
```

### Bar Charts

```python
# Plotting the bar chart

plt.figure(figsize=(12, 6))
cutCounts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Diamond Cuts')
plt.xlabel('Cut Type')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
# Plotting the bar chart

plt.figure(figsize=(12, 6))
colorCounts.plot(kind='bar', color='lightgreen')
plt.title('Distribution of Diamond Colors')
plt.xlabel('Color Type')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
# Plotting the bar chart

plt.figure(figsize=(12, 6))
clarityCounts.plot(kind='bar', color='lightpink')
plt.title('Distribution of Diamond Clarity')
plt.xlabel('Clarity Type')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--')
plt.show()
```

### Pie Charts

```python
# Plotting the pie chart

plt.figure(figsize=(10, 6))
cutCounts.plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightgreen', 'orange', 'lightcoral', 'gold'])
plt.title('Distribution of Diamond Cuts')
plt.ylabel('') 
plt.show()
```

```python
# Plotting the pie chart

plt.figure(figsize=(10, 6))
colorCounts.plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightgreen', 'orange', 'lightcoral', 'gold','cyan','red'])
plt.title('Distribution of Diamond Colors')
plt.ylabel('') 
plt.show()
```

```python
# Plotting the pie chart

plt.figure(figsize=(10, 6))
clarityCounts.plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightgreen', 'orange', 'lightcoral', 'gold','cyan','red','blue'])
plt.title('Distribution of Diamond Clarity')
plt.ylabel('') 
plt.show()
```
