### Handling Missing Values

```python
# to drop the Id column which is unnecssary for the dataset

df.drop('Id', axis= 1, inplace = True)
```

```python
# to view all the columns present in the dataset

df.columns
```

```python
# To drop the columns that has data missing greater than 20 %

threshold = len(df)*0.20
column = df.isna().sum() > threshold
columnToBeDropped = df.columns[column]
```

```python
# To know how many columns follow the above condion

columnToBeDropped.shape
```

```python
# To drop the columns that all follow the above condition

df.drop(columnToBeDropped, axis= 1, inplace = True)
df.shape
```

```python
# To drop the columns due to low variability

df.drop(
    [
        "Street",
        "Utilities",
        "Condition2",
        "RoofMatl",
        "ExterCond",
        "BsmtCond",
        "Heating",
        "GarageCond",
    ],
    axis=1,
    inplace=True,
)
```

```python
# To know the shape of the dataframe after removing certain columns

df.shape
```

```python
# To replace the  missing data in numeric columns by median

for col in df.select_dtypes(include=['int64','float64']):
    if df[col].isnull().sum() > 0:
        median_value = df[col].median()
        df[col].fillna(median_value, inplace=True)
```

```python
# To check how many null values exist in the dataframe after performing the above replacement opeation

df.isnull().sum()
```

```python
# To Handle missing data in categorical columns by mode

for col in df.select_dtypes(include=['object']):
    if df[col].isnull().sum() > 0:
        mode_value = df[col].mode()[0]
        df[col].fillna(mode_value, inplace=True)
```

```python
# To check whether we have any null values present or not

df.isnull().sum()
```

### Data Transformation & Reduction & Feature Engineering

```python
# To combine the both column to one

df["Total_floors"] = df["1stFlrSF"] + df["2ndFlrSF"]
df["Total_floors"].head()
```

```python
# To combine the both column to one

df["House_quality_score"] = df["OverallCond"] * df["OverallQual"]
df["House_quality_score"].head()
```

```python
# To combine the three column to one

df["Total_porch_area"] = (
    df["OpenPorchSF"] + df["EnclosedPorch"] + df["3SsnPorch"] + df["ScreenPorch"]
)
df["Total_porch_area"].head()
```

```python
# To combine the both column to one

df["TotalBsmtFinSF"] = df["BsmtFinSF1"] + df["BsmtFinSF2"]
```

```python
# To combine the both column to one

df["TotalFullBath"] = df["BsmtFullBath"] + df["FullBath"]
df["TotalHalfBath"] = df["BsmtHalfBath"] + df["HalfBath"]
```

```python
# # To drop the irrelevant columns

df.drop(
    [
        "BsmtFullBath",
        "FullBath",
        "BsmtHalfBath",
        "HalfBath",
        "OpenPorchSF",
        "EnclosedPorch",
        "3SsnPorch",
        "ScreenPorch",
        "OverallCond",
        "OverallQual",
        "1stFlrSF",
        "2ndFlrSF",
    ],
    axis=1,
    inplace=True,
)
```

### Encoding Categorical Columns

```python
# To load the Preprocessing

from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()

# To iterate over all columns in the DataFrame

for col in df.select_dtypes(include=['object']).columns:
    df[col] = label_encoder.fit_transform(df[col])
```
